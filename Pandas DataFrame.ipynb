## Pivot Table

```python
data = pd.read_csv(file_path)
data.head(2)
```

```python
data.pivot_table(index = 'Sex', values= 'Survived')
```

```python
data.pivot_table(index='Sex', values = 'Survived', aggfunc = [len, sum, 'mean'])
```

```python
data.pivot_table(index='Sex',columns='Pclass',values = 'Survived', aggfunc = [len, sum, 'mean'])
```

```python
data.pivot_table(index='Sex',
                 columns=['Embarked','Pclass'],
                 values = 'Survived', 
                 aggfunc = [len, sum, 'mean'])
```

```python
data.pivot_table(index=['Embarked','Sex'],
                 columns='Pclass',
                 values = 'Survived', 
                 aggfunc = [len, sum, 'mean'])
```

```python
data.pivot_table(index=['Embarked','Sex'],
                 columns ='Pclass',
                 values = 'Survived', 
                 aggfunc = [len, sum, 'mean'],
                 margins = True)
```

## Group By

```python
data.groupby(['Sex','Pclass'])['Survived'].count()
```

```python
data.groupby(['Sex','Pclass'])['Survived'].mean()
```

```python
data.groupby(['Sex','Pclass'])['Survived'].agg([len,sum,'mean'])
```

```python
data.groupby(['Sex','Pclass'])['Survived','Age'].agg({'Survived':[len,sum,'mean'], 'Age':'mean'})
```

```python
data.groupby(['Sex','Pclass'])['Survived','Age'].agg({'Survived': sum, 'Age':'mean'})
```

```python
data.groupby(['Sex','Pclass'])['Survived','Age'].agg({'Survived': sum, 
        'Age':'mean'}).stack().reset_index()#.rename(columns={'level_2':'Category', 0:'Result'})
```

```python
new_data = data.groupby(['Sex','Pclass'])['Survived','Age'].agg({'Survived': sum, 
        'Age':'mean'}).stack().reset_index().rename(columns={'level_2':'Category', 0:'Result'})
```

```python
new_data.head()
```

```python
new_data.sort_values('Category')
```
